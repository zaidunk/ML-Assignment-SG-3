```python
import pandas as pd

url = "https://raw.githubusercontent.com/zaidunk/ML-Assignment-SG-3/refs/heads/main/phone_usage_india.csv"

pui = pd.read_csv(url)
print(pui.columns)
```

Import data Phone Usage India

```python
import matplotlib.pyplot as plt
pui.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
pui.plot(kind="scatter", x="Age", y="Screen_Time_(hrs/day)", alpha=0.1)
```

Berdasarkan scatter plot tersebut, penyebaran screen time merata pada semua usia.



```python
corr_matrix = pui.select_dtypes(include=['number']).corr()
corr_matrix["Age"].sort_values(ascending=False)
```

Dapat kita lihat di atas, korelasi antara usia pengguna dengan variabel yang lain sangatlah rendah.

```python
from pandas.plotting import scatter_matrix

attributes = ["Age", "Screen_Time_(hrs/day)", "Data_Usage_(GB/month)",
              "Calls Duration_(mins/day)","Number_of_Apps_Installed","Social_Media_Time_(hrs/day)","E-commerce_Spend_(INR/month)","Streaming_Time_(hrs/day)","Gaming_Time_(hrs/day)","Monthly_Recharge_Cost_(INR)"]
scatter_matrix(pui[attributes], figsize=(20, 10), alpha=0.1)
```
